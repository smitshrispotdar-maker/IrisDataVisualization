```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
iris = load_iris(as_frame=True)

df = iris.frame
```

```python
df["species"] = df["target"].map(dict(enumerate(iris.target_names)))
df.drop("target", axis=1, inplace=True)

print("First 5 Rows")
print(df.head())

print("\nDataset Shape:")
print(df.shape)

print("\nInformation:")
print(df.info())

print("\nStatistics:")
print(df.describe())
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(
    data=df,
    x="sepal length (cm)",
    y="sepal width (cm)",
    hue="species",
    s=100
)
plt.title("Sepal Length vs Sepal Width")
plt.show()
```

```python
pair = sns.pairplot(df, hue="species")
pair.fig.suptitle("Pair Plot of Iris Dataset", y=1.02)

plt.show()
```

```python
df.hist(figsize=(12,8), bins=15)
plt.suptitle("Feature Distributions")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(
    data=df,
    x="species",
    y="petal length (cm)"
)
plt.title("Petal Length by Species")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.violinplot(
    data=df,
    x="species",
    y="petal width (cm)"
)
plt.title("Petal Width Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,6))

corr = df.iloc[:, :4].corr()

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm",
    linewidths=0.5
)

plt.title("Correlation Heatmap")
plt.show()
```
